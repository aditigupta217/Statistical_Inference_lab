```python
data=soup.find_all('div')
print(data)
```

```python
printed = set()
for d in data:
  meta_tag = d.find("meta")

  if meta_tag is not None:
    content = meta_tag.get("content")

    if content:

      if content not in printed:
        print(content)
        printed.add(content)
```

```python
import pandas as pd
data = soup.find_all('div',class_="quote")
Quotes=[]
for D in data:
  Quote = D.find('span',class_="text").text
  Quotes.append(Quote)
  print(Quote)
```

```python
import pandas as pd

Quotes = []
Authors = []
Tags = []

for D in data:
    # Extract Quote
    quote = D.find('span', class_="text").text
    Quotes.append(quote)

    # Extract Author
    author = D.find('small', class_="author").text
    Authors.append(author)

    # Extract Tags
    tag = D.find("meta")["content"]
    Tags.append(tag)

# Create DataFrame after the loop
x = pd.DataFrame({
    "Quotes": Quotes,
    "Authors": Authors,
    "Tags": Tags
})


```
