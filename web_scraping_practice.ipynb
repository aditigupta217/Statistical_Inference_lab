```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python

webpage = requests.get('https://realpython.github.io/fake-jobs/?utm_source=chatgpt.com$0')
```

```python
print(type(webpage))
```

```python
soup = BeautifulSoup(webpage.text, "lxml")
```

```python
print(soup.prettify())
```

```python
soup.find_all('h1')[0].text.strip()
```

```python
soup.find_all('h2')
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
for i in soup.find_all('h3'):
    print(i.text.strip())
```

```python
for i in soup.find_all('p', class_ = 'location'):
    print(i.text.strip())
```

```python
company = soup.find_all('div', class_ = 'card-content')
```

```python
len(company)
```

```python
name = []
subtitle = []
location = []
date = []

for i in company:
     name.append(i.find('h2').text.strip())
     subtitle.append(i.find('h3').text.strip())
     location.append(i.find('p', class_ = 'location').text.strip())
     date.append((i.find('time')['datetime']))
d = {'name':name , 'subtitle':subtitle, 'location': location,'date' : date}
df = pd.DataFrame(d)
df.to_csv('jobs.csv')

```
