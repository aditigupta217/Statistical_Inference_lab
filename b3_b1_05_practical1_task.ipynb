```python
!pip install requests
import requests
url = "https://books.toscrape.com/"
response = requests.get('https://books.toscrape.com/')
```

```python
print(response.status_code)
```

```python

from bs4 import BeautifulSoup
```

```python
soup=(BeautifulSoup(response.content,'html.parser'))
print(BeautifulSoup)
```

```python
print(soup.prettify())
```

```python
import pandas as pd
```

```python
print(soup.find('title'))
```

```python
books = []
for book in soup.find_all("article", class_="product_pod"):
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    rating = book.find("p", class_="star-rating")["class"][1]

    books.append({
        "Book Name": title,
        "Price": price,
        "Rating": rating
    })
df = pd.DataFrame(books)
print(df)
```

```python
images = []

for book in soup.find_all("article", class_="product_pod"):
    image = book.find("img")["src"]
    image_url = "https://books.toscrape.com/" + image.replace("../", "")
    images.append(image_url)

df["Image"] = images

print(df)
```

```python
url = "https://www.scrapethissite.com/pages/simple/"
response = requests.get('https://www.scrapethissite.com/pages/simple/')
```

```python
soup=(BeautifulSoup(response.content,'html.parser'))
print(BeautifulSoup)
```

```python
print(soup.find('title'))
```

```python
countries = soup.find_all("div", class_="country")

country_list = []
capital_list = []
population_list = []
area_list = []

for country in countries:
    country_list.append(country.find("h3", class_="country-name").get_text(strip=True))
    capital_list.append(country.find("span", class_="country-capital").get_text(strip=True))
    population_list.append(country.find("span", class_="country-population").get_text(strip=True))
    area_list.append(country.find("span", class_="country-area").get_text(strip=True))

df = pd.DataFrame({
    "Country": country_list,
    "Capital": capital_list,
    "Population": population_list,
    "Area": area_list
})

print(df)
```

```python
url = "https://www.scrapethissite.com/pages/forms/"
response = requests.get('https://www.scrapethissite.com/pages/forms/')
```

```python
soup=(BeautifulSoup(response.content,'html.parser'))
print(BeautifulSoup)
```

```python
print(soup.find('title'))
```

```python
rows = soup.find_all("tr", class_="team")

team_name = []
year = []
wins = []
losses = []
ot_losses = []
win_pct = []

for row in rows:
    team_name.append(row.find("td", class_="name").get_text(strip=True))
    year.append(row.find("td", class_="year").get_text(strip=True))
    wins.append(row.find("td", class_="wins").get_text(strip=True))
    losses.append(row.find("td", class_="losses").get_text(strip=True))
    ot_losses.append(row.find("td", class_="ot-losses").get_text(strip=True))
    win_pct.append(row.find("td", class_="pct").get_text(strip=True))

df = pd.DataFrame({
    "Team Name": team_name,
    "Year": year,
    "Wins": wins,
    "Losses": losses,
    "OT Losses": ot_losses,
    "Win %": win_pct
})

print(df)
```

```python
df.to_csv("hockey.csv")
```

```python
url = "https://nitrr.ac.in/aboutcse.php"
```

```python
response = requests.get(url,verify=False)
```

```python
soup=(BeautifulSoup(response.content,'html.parser'))
print(BeautifulSoup)
```

```python
print(soup.find('title'))
```

```python
faculty = soup.find_all("div", class_="faculty")

faculty_data = []

for person in faculty:
    name = person.find("h4").text.strip()
    designation = person.find_all("p")[0].text.strip()
    qualification = person.find_all("p")[1].text.strip()

    faculty_data.append({
        "Name": name,
        "Qualification": qualification,
        "Designation": designation
    })

df = pd.DataFrame(faculty_data)

print(df)
```
